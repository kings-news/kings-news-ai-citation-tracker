```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/citation_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['ai_citation_score','search_visibility_score','citation_readiness_score','visibility_gap_score','platform_reach_score','content_optimization_score']
signal_avgs = df[signal_cols].mean()
colors = ['#6366f1','#3b82f6','#10b981','#ef4444','#f59e0b','#8b5cf6']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('Kings News - Average AI Citation Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Citation Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=70, color='green', linestyle='--', alpha=0.5, label='Healthy threshold')
plt.legend()
plt.tight_layout()
plt.show()
print(signal_avgs.round(1).to_string())
```

```python
platform_cols = ['chatgpt_score','google_ai_score','gemini_score','perplexity_score','copilot_score']
platform_avgs = df[platform_cols].mean()
plt.figure(figsize=(10,5))
plt.bar(platform_avgs.index, platform_avgs.values, color=['#10b981','#3b82f6','#6366f1','#f59e0b','#8b5cf6'])
plt.title('Kings News - Average Platform Citation Scores', fontsize=14, fontweight='bold')
plt.xlabel('AI Platform')
plt.ylabel('Average Score (0-100)')
plt.tight_layout()
plt.show()
print(platform_avgs.round(1).to_string())
```
